```python
from selenium import webdriver 
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.service import Service
```

```python
service = Service(executable_path="geckodriver.exe")
```

```python
def init_driver(service):
    
    firefox_profile = webdriver.FirefoxProfile()
    
    firefox_profile.set_preference('dom.ipc.plugins.enabled.libflashplayer.so', False)  # close the flashplayer
    firefox_profile.set_preference("media.volume_scale", "0.0")  # close the volume of notification
    firefox_profile.set_preference("dom.webnotifications.enabled", False) # close the notification

    options = Options()
    options.profile = firefox_profile # to close the image loading
    return driver
driver = webdriver.Firefox(service=service)
```

```python
all = driver.find_elements(By.CSS_SELECTOR , 'div.-paxs.row._no-g._4cl-3cm-shs article.prd._fb.col.c-prd')
len(all)                                       
```

```python
title_list = []
for product in all:
    title = product.find_element(By.CSS_SELECTOR, 'div.info h3.name').text
    title_list.append(title)
    print(title)
```

```python

price_list = []
for price in all:
    price1 = price.find_element(By.CSS_SELECTOR, 'div.info div.prc').text
    price_list.append(price1)
    print(price1)
```

```python
import pandas as pd
df = pd.DataFrame(title_list , columns=["Product"])
df
```

```python
df['Price after sale'] = price_list
```

```python
df
```

```python
oldprice_list = []
for price2 in all:
    try:
        price3 = price2.find_element(By.CSS_SELECTOR, 'div.info div.old').text
        oldprice_list.append(price3)
    except:
        oldprice_list.append(None)
    print(price3)
```

```python
df["Price"] = oldprice_list
df
```

```python
url_list = []
for product in all:
    product_title = product.find_element(By.CSS_SELECTOR, 'div.info h3.name').text
    # print(product_title)
    product_url = product.find_element(By.CSS_SELECTOR, 'a.core').get_attribute('href')
    url_list.append(product_url)
    print(product_url)
```

```python
df["URL"] = url_list
df
```
